# Now will scrap out the data from all 559 stores.

```python
store_name_lst=[]
city_name_lst=[]
full_address_lst=[]
phone_lst=[]
timing_lst=[]
coordinate_lst=[]
for i in range(len(response)):
    z=BeautifulSoup(response[i], 'html.parser')
    for j in range(len(z.find_all('div',class_='card h-100 info-card idx-str-info-card rounded-0'))):
        
        store_name_lst.append(z.find_all('div',class_='card h-100 info-card idx-str-info-card rounded-0')[j].find('a',class_='col-store-name txt-color-3 txt-hover-1').text)
        
        city_name_lst.append(z.find_all('div',class_='card h-100 info-card idx-str-info-card rounded-0')[j].find('span',class_='mb-0 idx-str-locality').text.replace('\xa0',''))
        
        full_address_lst.append(z.find_all('div',class_='card h-100 info-card idx-str-info-card rounded-0')[j].find('span',class_='store-address-info idx-info-card-str-add-info').text)
        
        phone_lst.append(z.find_all('div',class_='card h-100 info-card idx-str-info-card rounded-0')[j].find('a',class_='txt-hover-1 txt-color-1').text)
        
        timing_lst.append(z.find_all('div',class_='card h-100 info-card idx-str-info-card rounded-0')[j].find('span',class_='store-time-info idx-info-card-str-time-info').text)
        
        coordinate_lst.append(z.find_all('div',class_='col-md-6 col-sm-6 col-6 form-group mb-0 pl-2')[j].find('a')['href'].split('=')[-1])
    
    
    
    
```

```python
Vishal_store_df=pd.DataFrame(data={'store_name':store_name_lst,'city_name':city_name_lst,'full_address':full_address_lst
                                   ,"Coordinates(Longtude/Latitude)":coordinate_lst,'phone_no':phone_lst,
                                   'store_timing':timing_lst})
```
